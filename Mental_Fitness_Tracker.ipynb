```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
  if df[i].dtype=='object':
    df[i]=l.fit_transform(df[i])
```

```python
df.shape
```

# Spliting the Data and Model Building

```python
X=df.drop('mental_fitness',axis=1)
y=df['mental_fitness']

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.20,random_state=2)
```

```python
print("xtrain:",xtrain.shape)
print("xtest:",xtest.shape)
print("ytrain:",ytrain.shape)
print("ytest:",ytest.shape)

```

```python
#LINEAR REGRESSION

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

lr=LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
#model evaluation

ytrain_pred=lr.predict(xtrain)
mse=mean_squared_error(ytrain,ytrain_pred)

rmse=(np.sqrt(mse))

r2=r2_score(ytrain,ytrain_pred)
```

```python
print("The Linear Regression model performance for training set")
print("**********************************")
print("MSE is {}".format(mse))
print("RMSE is {}".format(rmse))
print("R2 SCORE is {}".format(r2))
```

```python
#RANDOM FOREST REGRESSOR

from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor()
rf.fit(xtrain,ytrain)
```

```python
#model evaluation

ytrain_pred=rf.predict(xtrain)
mse1=mean_squared_error(ytrain,ytrain_pred)

rmse1=(np.sqrt(mse))

r21=r2_score(ytrain,ytrain_pred)
```

```python
print("The Random Forest Regressor model performance for training set")
print("**********************************")
print("MSE is {}".format(mse1))
print("RMSE is {}".format(rmse1))
print("R2 SCORE is {}".format(r21))
```
